```
import matplotlib.pyplot as plt
import numpy as np
import random
```

##Histogram

```
miet_student_age=np.random.randint(18,45,(100))
mit_student_age=np.random.randint(15,40,(100))
```

* if you want to draw histogram 
* Then use matplotlib's hist()

```
plt.hist(miet_student_age)
plt.show()
```

* But this histogram is not looking good
* Because there is no xlable,  y lable and title
* For this purpose we will use
  

1.   xlabel
2.   ylabel
3.   title


* so let's try for this

```
plt.hist(miet_student_age)
plt.title("MIET student's histogram",fontsize=15)
plt.xlabel("Student age")
plt.ylabel("No.of student")
plt.show()
```

* Let's understand parameter of hist()
one by one
* The first parameter is data which is 'miet_student_age' in our case
* Now understand bins means we have to define category of each BAR in histogram
* we are defines bins(according your data)

```
bins=[15,20,25,30,40,45]
plt.hist(miet_student_age,bins)
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.show()
```

* If you observe the histogram 
* It takes 100% sapce or you can say there is no gap b/w the bar
* So now we we will take 80% for bar and 20% for gap b/w each bar
* Default is 100% you can take any value

```
  bins=[15,20,25,30,40,45]
plt.hist(miet_student_age,bins,rwidth=0.8)
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.show()
```

* Next type is histtype
* we can set histtype value as 'bar', 'barstacked', 'step', 'stepfilled'
* you can take any value and check the output

```
#histtype---step
plt.hist(miet_student_age,bins,rwidth=0.8,histtype='step')
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.show()
```

```
#histtype
plt.hist(miet_student_age,bins,rwidth=0.8,histtype='bar')
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.show()
```

* Next parameter is orientation
* it can take 'horizontal', 'vertical'

```
#histtype
plt.hist(miet_student_age,bins,rwidth=0.8,orientation='horizontal')
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.show()
```

* Note there is a parameter align which can be 'left','mid','right'
* you should check by yourself
* Here we skip this parameter

* Next parameter is color
* if you want to change the color of your bar
* Then use color parameter

```
#histtype
plt.hist(miet_student_age,bins,rwidth=0.8,color='r')
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.show()
```

* Next parameter is label
* when you want to set label in your histogram then use lable parameter
* BUT label parameter is not working without lenged() 
* so you need to use legend() with label

```
#histtype
plt.hist(miet_student_age,bins,rwidth=0.8,color='r',label="miet student")
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no of student")
plt.legend()
plt.show()
```

* we can do use legend for the same work as above 
* so no need to use label parameter

```
#histtype
plt.hist(miet_student_age,bins,rwidth=0.8,color='r')
plt.title("miet student's histogram")
plt.xlabel("student age")
plt.ylabel("no.of student")
plt.legend(["age of student"],loc=2)
plt.show()
```

### Two or more histogram in one 

```
miet_student_age=np.random.randint(18,45,(100))
mit_student_age=np.random.randint(15,40,(100))
```

```
#histtype
plt.hist(miet_student_age,rwidth=0.8,color='r',label="miet student")
plt.hist(mit_student_age,rwidth=0.8,color='y',label="mit student")
plt.title("miet and mit student's histogram")
plt.xlabel("student age")
plt.ylabel("no of student")
plt.legend(loc=2)
plt.show()
```

* BUt the above histogram is overlapping
* and we don't want this so take both data miet_student_age,mit_student_age
* Remember this must be taken in LIST as below
* We can take color and label in LIST also

```
#histtype
plt.hist([miet_student_age,mit_student_age],rwidth=0.8,color=['r','y'],label=["miet student","mit students"])

plt.title("miet and mit student's histogram")
plt.xlabel("student age")
plt.ylabel("no of student")
plt.legend(loc=2)
plt.show()
```

* If you need to increase the size of figure size
* Then use figure() and give the parameter figsize according your need


```
#histtype
plt.figure(figsize=(16,9))
plt.hist([miet_student_age,mit_student_age],rwidth=0.8,color=['r','y'],label=["miet student","mit students"])

plt.title("miet and mit student's histogram")
plt.xlabel("student age")
plt.ylabel("no of student")
plt.legend(loc=2)
plt.show()
```

* If you need grid in background
* use style module 
* And style.use('ggplot')

```
from matplotlib import style
```

```
#histtype
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.hist([miet_student_age,mit_student_age],rwidth=0.8,color=['r','y'],label=["miet student","mit students"])

plt.title("miet and mit student's histogram")
plt.xlabel("student age")
plt.ylabel("no of student")
plt.legend(loc=2)
plt.show()
```

* To change color font , width then use grid()

```
#histtype
style.use('ggplot')
plt.figure(figsize=(16,9))
plt.hist([miet_student_age,mit_student_age],rwidth=0.8,color=['r','y'],label=["miet student","mit students"])
plt.grid()# you can change color and others parameters in grid ()
plt.title("miet and mit student's histogram")
plt.xlabel("student age")
plt.ylabel("no of student")
plt.legend(loc=2)
plt.show()
```
