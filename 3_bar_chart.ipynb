```
import matplotlib.pyplot as plt
```

```
classes=['py','R','Ai','Ml','DS']
class_1=[30,10,20,25,10]
class_2=[40,5,20,20,10]
class_3=[35,5,30,15,15]
```

### BAR Chart

```
#bar()
plt.bar(classes,class_1)
plt.show()
```

* If you need to work on horizontal bar
* Then use barh() instead of bar()

```
plt.barh(classes,class_1)
plt.show()
```

* if you need to chhange the width
* Then use width parameter

```
plt.bar(classes,class_1,width=0.5)
plt.show()
```

* Next parameter is align
* there are two values we can use which are-'center', 'edge'
* Default is 'center

```
plt.bar(classes,class_1,width=0.5,align='edge')
plt.show()
```

* If you need to change the color of bar 
* Then use color parameter

```
plt.bar(classes,class_1,width=0.5,align='edge',color='g')
plt.show()
```

* If you want to color edge of the bar
* Then use edgecolor parameter

```
plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r')
plt.show()
```

* If you change the width of edge
* Then use linewidth parameter

```
plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5)
plt.show()
```

* if you need to change the opecity of colr
* Then use alpha parameter
* you can take value b/w 0 to 1

```
plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5,alpha=0.9)
plt.show()
```

* if you want to change the style of your bars 
* Then use linestyle
* '-', '--', '-.', ':', '', (offset, on-off-seq), ...
* you can take any value from above

```
plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5,alpha=0.9,linestyle='--',)
plt.show()
```

* visible parameter is True (default)
* if you make False then it will not visible

```
  plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5,alpha=0.9,linestyle='--',
        visible=False)
plt.show()
```

* Now if we need label on the graph
* Then use lable parameter 
* But use legend() also 

```
  plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5,alpha=0.9,linestyle='--',
        label='plottingbar')
  plt.legend()
plt.show()
```

* you can change the figuresize 
* by using plt.figure(figsize=())

```
plt.figure(figsize=(10,7))
plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5,alpha=0.9,linestyle='--',
        )
plt.show()
```

* If you need grid in your garph
* Then use style module
* and style.use('ggplot')

```
from matplotlib import style
```

```
plt.figure(figsize=(10,7))
style.use('ggplot')
plt.bar(classes,class_1,width=0.5,align='edge',color='y',edgecolor='r',linewidth=5,alpha=0.9,linestyle='--',
        label="plotting bar")
plt.title("bar chart")
plt.xlabel("classes")
plt.ylabel("no of students")
plt.legend()
plt.show()
```

### Two or more bar chart in one

```
plt.figure(figsize=(10,7))
style.use('ggplot')
plt.bar(classes,class_1,width=0.5,align='edge',color='y',label="CLASS 1")
plt.bar(classes,class_2,width=0.5,align='edge',color='m',label="CLASS 2")


plt.title("BAR CHART",fontsize =15)
plt.xlabel("classes")
plt.ylabel("no of students")
plt.legend()
plt.show()
```

```
import numpy as np
```

```
plt.figure(figsize=(10,7))
style.use('ggplot')

plt.bar(classes,class_1,align='edge',width=0.2,color='y',label="plotting bar")
plt.bar(classes,class_2,align='edge',width=0.2,color='m',label="plotting bar")
plt.bar(classes,class_3,align='edge',width=0.2,color='g',label="plotting bar")

plt.title("bar chart")
plt.xlabel("classes")
plt.ylabel("no of students")
plt.legend()
plt.show()
```

* in the above bar chart all the bar chart is overlapping
* we don't want such type of bar chart we need seprate bar chart

```
# here we need classes index like 0,1,2... 
# so we will use numpy
x=np.arange(len(classes))
x
```

```
plt.figure(figsize=(10,7))
style.use('ggplot')
width=0.2
plt.bar(x,class_1,width,align='edge',color='y',label="plotting bar")
plt.bar(x+width,class_2,width,align='edge',color='m',label="plotting bar")
plt.bar(x+width+width,class_3,width,align='edge',color='k',label="plotting bar")

plt.title("bar chart")
plt.xlabel("classes")
plt.ylabel("no of students")
plt.legend()
plt.show()
```

* if you observe the char in x axis there is no classes name
* so for that we have to use xticks()

```
plt.figure(figsize=(10,7))
style.use('ggplot')
width=0.2
plt.bar(x,class_1,width,align='edge',color='y',label="plotting bar")
plt.bar(x+width,class_2,width,align='edge',color='m',label="plotting bar")
plt.bar(x+width+width,class_3,width,align='edge',color='k',label="plotting bar")
plt.xticks(x+width,classes)
plt.title("bar chart")
plt.xlabel("classes")
plt.ylabel("no of students")
plt.legend()
plt.show()
```

* if you rotate the classes name 
* Then use rotation parameter in xticks

```
plt.figure(figsize=(10,7))
style.use('ggplot')
width=0.2
plt.bar(x,class_1,width,align='edge',color='y',label="plotting bar")
plt.bar(x+width,class_2,width,align='edge',color='m',label="plotting bar")
plt.bar(x+width+width,class_3,width,align='edge',color='k',label="plotting bar")
plt.xticks(x+width,classes,rotation=30)
plt.title("bar chart")
plt.xlabel("classes")
plt.ylabel("no of students")
plt.legend()
plt.show()
```

* for horizontal plotting

```
plt.figure(figsize=(10,7))
style.use('ggplot')
width=0.2
plt.barh(x,class_1,width,align='edge',color='y',label="plotting bar")
plt.barh(x+width,class_2,width,align='edge',color='m',label="plotting bar")
plt.barh(x+width+width,class_3,width,align='edge',color='k',label="plotting bar")
plt.yticks(x+width,classes,rotation=20)
plt.title("bar chart")
plt.ylabel("classes")
plt.xlabel("no of students")
plt.legend()
plt.show()
```
